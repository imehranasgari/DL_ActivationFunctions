```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# بازه (-10, 10)
x = np.linspace(-10, 10, 100)
```

```python
# توابع فعال‌ساز
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def elu(x, alpha=1):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

def selu(x):
    alpha = 1.67326324
    lambda_ = 1.05070098
    return lambda_ * np.where(x >= 0, x, alpha * (np.exp(x) - 1))

# مشتق توابع
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def relu_derivative(x):
    return np.where(x >= 0, 1, 0)

def elu_derivative(x, alpha=1):
    return np.where(x >= 0, 1, alpha * np.exp(x))

def selu_derivative(x):
    alpha = 1.67326324
    lambda_ = 1.05070098
    return lambda_ * np.where(x >= 0, 1, alpha * np.exp(x))
```

```python
# رسم توابع
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, tanh(x), label='Tanh')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, elu(x), label='ELU')
plt.plot(x, selu(x), label='SELU')
plt.title('Activation Functions')
plt.legend()
plt.grid(True)  # نمایش شبکه
plt.axhline(0, color='gray')  # محور y
plt.axvline(0, color='gray')  # محور x

plt.subplot(1, 2, 2)
plt.plot(x, sigmoid_derivative(x), label='Sigmoid Derivative')
plt.plot(x, tanh_derivative(x), label='Tanh Derivative')
plt.plot(x, relu_derivative(x), label='ReLU Derivative')
plt.plot(x, elu_derivative(x), label='ELU Derivative')
plt.plot(x, selu_derivative(x), label='SELU Derivative')
plt.title('Derivatives of Activation Functions')
plt.legend()

plt.tight_layout()
plt.grid(True)  # نمایش شبکه
plt.axhline(0, color='gray')  # محور y
plt.axvline(0, color='gray')  # محور x
plt.show()
```
